SERIES

```python
#ITS A ONE - DIMENTIONAL ARRAY
#IT CAN STORE ANY DATA TYPE
#THERE ARE MANY TYPES TO CREATE SEREIES
```

```python
import numpy as np
import pandas as pd
```

```python
lables=['a','b','c']
my_list=[10,20,30]
di={1:11,2:22,3:33}
arr=np.array([10,20,30])
```

```python
#SERIES USING LIST
li=pd.Series(my_list,lables)
li
```

```python
#SERIES USING ARRAY
ar=pd.Series(arr,lables)
ar
```

```python
#SERIES USING DICTONARY
dit=pd.Series(di)
dit
```
